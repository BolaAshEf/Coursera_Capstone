```python
# For some Reason I decided to replace this weired symbol(	) to this symbol(~) and continue
List_of_postal_codes_of_Canada_seperated = List_of_postal_codes_of_Canada_str.replace("	", "~")
```

```python
# Get an array of Row (by splitting string at (\n))
List_of_postal_codes_of_Canada_rows = List_of_postal_codes_of_Canada_seperated.split("\n")
# List_of_postal_codes_of_Canada_rows
```

```python
# First I create a dict to hold columns and rows
postal_codes_of_Canada_dict = {'PostalCode':[], 'Borough':[], 'Neighborhood':[]}

# Loop over each row and get columns by splitting string at(~) and fill the above dict
for row in List_of_postal_codes_of_Canada_rows:
    cols = row.split('~')
    postal_code = cols[0]
    borough = cols[1]
    neighborhood = cols[2]
    
    # Check if the borough of this row is 'Not assigned' if True then discard it
    if borough == 'Not assigned':
        continue
    # Check if the neighborhood of this row is 'Not assigned' if True then assign the borough value to it
    if neighborhood == 'Not assigned':
        borough = neighborhood
    
    # After checking fill the dict with data 
    postal_codes_of_Canada_dict['PostalCode'].append(postal_code)
    postal_codes_of_Canada_dict['Borough'].append(borough)
    postal_codes_of_Canada_dict['Neighborhood'].append(neighborhood)
    
# Just for confirm that all lengths of the rows of each column are equal
print(len(postal_codes_of_Canada_dict['PostalCode']))
print(len(postal_codes_of_Canada_dict['Borough']))
print(len(postal_codes_of_Canada_dict['Neighborhood']))
```
